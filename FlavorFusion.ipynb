```python
def proper_df(df):

  df.columns=[i.lower().replace(' (₹)','').replace(' ','_').replace('.','').replace('@','')\
              .replace('!','').replace('#','').replace('$','').replace('%','').replace('&','')\
              .replace('(','').replace(')','') for i in df.columns]

  df.drop([i for i in df.columns if df[i].isna().sum()==len(df)],axis=1,inplace=True)

  newdf = pd.DataFrame()

  dict1={i:df[i].isna().sum() for i in df.columns if df[i].isna().sum()!=0}

  newdf['colname'] = dict1.keys()
  newdf['nullcount'] = dict1.values()
  newdf['valuecounts'] = [df[i].value_counts().to_dict() for i in dict1.keys()]

  return newdf

  
```

```python
menu_list = []
for i in menu.keys():
  for j in menu[i]:
    menu_list.append(j)


menu_list
```

```python
import difflib

correct_items = []

for i in df['items']:

  words = i.split(',')

  corrected_words = []
  for word in words:
    # Find the closest match for each word in the food list
    closest_match = difflib.get_close_matches(word, menu_list, n=1)
    if closest_match:
        corrected_words.append(closest_match[0])
    else:
        corrected_words.append(word)


  corrected_string = ','.join(corrected_words)

  correct_items.append(corrected_string)



correct_items
```

```python
df['items1'] = correct_items
```

```python
df['itemslist'] = df['items1'].apply(lambda x: x.split(','))
df['itemslist']
```
